# Ridge Regression and Lasso

This notebook explores ridge regression and lasso. These alternative linear fitting techniques can improve a model's performance and interpretability.

## Import libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

%matplotlib inline
```

## Exploratory Data Analysis

## Preprocessing

### Encoding all categorical feature

```python
categorical_columns = ['Brand', 'Tipe', 'CPU', 'RAM', 'Memori', 'GPU', 'SistemOperasi', 'Layar']
for column in categorical_columns:
    data[column] = data[column].fillna('Unknown')
```

```python
data['Berat'] = data['Berat'].str.extract('(\d+\.?\d*)').astype(float) # Extract numbers including decimals
```

```python
df = pd.DataFrame(data)

# Applying one-hot encoding
data = pd.get_dummies(df, columns=['Brand', 'Tipe', 'CPU', 'RAM', 'Memori', 'GPU', 'SistemOperasi', 'Layar'])
```

```python
data.info()
```

## Modelling 

### Multiple linear regression - least squares fitting 

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

Xs = data.drop(['Harga'], axis=1)
y = data['Harga'].values.reshape(-1,1)

lin_reg = LinearRegression()

lin_reg.fit(Xs, y)

neg_MSEs = cross_val_score(lin_reg, Xs, y, scoring='neg_mean_squared_error', cv=5)

mean_MSE = -np.mean(neg_MSEs)

# Calculate the RMSE scores
rmse_scores = np.sqrt(mean_MSE)

print(rmse_scores)
```

### Ridge regression 

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge

alpha = [1e-15, 1e-10, 1e-8, 1e-4, 1e-3,1e-2, 1, 5, 10, 20]

ridge = Ridge()

parameters = {'alpha': [1e-15, 1e-10, 1e-8, 1e-4, 1e-3,1e-2, 1, 5, 10, 20]}

ridge_regressor = GridSearchCV(ridge, parameters,scoring='neg_mean_squared_error', cv=5)

ridge_regressor.fit(Xs, y)
```

```python
ridge_regressor.best_params_
```

```python
ridge_regressor.best_score_
```

```python
neg_mse_scores1 = cross_val_score(ridge_regressor, Xs, y, scoring='neg_mean_squared_error', cv=5)
mse_scores = -neg_mse_scores1
rmse_scores = np.sqrt(mse_scores)

print(rmse_scores)
```

### Lasso 

```python
from sklearn.linear_model import Lasso

lasso = Lasso()

parameters = {'alpha': [1e-15, 1e-10, 1e-8, 1e-4, 1e-3,1e-2, 1, 5, 10, 20]}

lasso_regressor = GridSearchCV(lasso, parameters, scoring='neg_mean_squared_error', cv = 5)

lasso_regressor.fit(Xs, y)
```

```python
lasso_regressor.best_params_
```

```python
lasso_regressor.best_score_
```

```python
neg_mse_scores2 = cross_val_score(lasso_regressor, Xs, y, scoring='neg_mean_squared_error', cv=5)
mse_scores = -neg_mse_scores2
rmse_scores2 = np.sqrt(mse_scores)
```

```python
print(rmse_scores2)
```
